```python
import numpy as np
import matplotlib.pyplot as plt
import vizdoom as vsd
```

# Brute Force Experiment

```python
epochs = 7
epochs = np.arange(1, epochs+1, 1)

file_dir = "Experiment_BRUTEFORCE/results/"

#Results Format: (4,5)
#[ [Mean Reward],[Standard Deviation],[Min Reward],[Max Reward] ]
Experiment_BRUTEFORCE_basic_upgraded_Rewards = np.load(file_dir + "basic_upgraded_Rewards.npy")

#----------------------------------------------------------------------------------------------------
fig, axs = plt.subplots(1, 2, figsize=(15,7))
axs[0].plot(epochs, Experiment_BRUTEFORCE_basic_upgraded_Rewards[0,:], 'tab:blue', label='Brute Force')
axs[0].set_title('Mean Reward')
axs[0].legend(loc='best')
axs[1].plot(epochs, Experiment_BRUTEFORCE_basic_upgraded_Rewards[1,:], 'tab:blue', label='Brute Force')
axs[1].set_title('Mean Standard Deviation')
axs[1].legend(loc='best')
for ax in axs.flat:
    ax.set(xlabel='Epoch (2000 steps each)', ylabel='Cumulative Reward')

#----------------------------------------------------------------------------------------------------
fig, axs = plt.subplots(1, 2, figsize=(15,7))
axs[0].plot(epochs, Experiment_BRUTEFORCE_basic_upgraded_Rewards[2,:], 'tab:blue', label='Brute Force')
axs[0].set_title('Mean Minimum Reward')
axs[0].legend(loc='best')
axs[1].plot(epochs, Experiment_BRUTEFORCE_basic_upgraded_Rewards[3,:], 'tab:blue', label='Brute Force')
axs[1].set_title('Mean Maximum Reward')
axs[1].legend(loc='best')
for ax in axs.flat:
    ax.set(xlabel='Epoch (2000 steps each)', ylabel='Cumulative Reward')
```
